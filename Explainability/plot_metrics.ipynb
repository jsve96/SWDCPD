```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loss_train = pd.read_csv("C:/Users/jsve/Downloads/Loss_train_loss.csv")
loss_val = pd.read_csv("C:/Users/jsve/Downloads/Loss_val_loss.csv")

acc_train = pd.read_csv("C:/Users/jsve/Downloads/Accuracy_train_acc.csv")
acc_val = pd.read_csv("C:/Users/jsve/Downloads/Accuracy_val_acc.csv")
```

```python
fig, ax = plt.subplots(1,2,figsize=(8,3))
loss_train['Value'].plot(marker='.',ax=ax[0],color='blue',label='train')
loss_val['Value'].plot(marker='.',ax=ax[0],color='red',label='val')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Loss')
ax[0].set_xticks([0,5,10,15])
ax[0].legend()



acc_train['Value'].plot(marker='.',ax=ax[1],alpha=1,color='blue',label='train')
acc_val['Value'].plot(marker='.',ax=ax[1],alpha=1,color='red',label='val')
ax[1].set_ylabel('Accuracy')
ax[1].set_xlabel('Epochs')
ax[1].legend()
ax[1].set_xticks([0,5,10,15])

fig.tight_layout()
```
